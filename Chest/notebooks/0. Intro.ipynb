Ya observamos que la imagenes estan compuestas de 0-255, dado esto vamos utilizar el metodo de ImageDateGenerator para variar de cierta forma mis imagenes y mejorar al modelo de entrenamiento neuronal

```python
train_datagen=ImageDataGenerator(rescale=1/255, vertical_flip=True, fill_mode='reflect')
```

```python
train_gen=train_datagen.flow_from_directory(
    "E:/piere/Documentos/Portafolio/torax/Data/RAW/Data/train",
    target_size=(200,200),
    batch_size=128,
    color_mode='grayscale',
    subset='training')
```

```python
plt.imshow(train_gen[0][0][6], cmap='Greys')
```
