```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
```

```python
df=pd.read_excel('e://piere/documentos/Portafolio/Oil Play Prediction/Data/RAW/2019 Atlas Update.xlsx')
```

```python
data=pd.concat([df["THK"],df["WDEP"],df["TVOL"],df["TAREA"],df["Pi"],df["POROSITY"],df["SW"],df["PERMEABILITY"],df["TI"],df["SS"],df["PLAY_TYPE"]],axis=1)
```

```python
per_nan=data[data["PERMEABILITY"].isnull()].index.tolist()
```

```python
data=data.drop(index=per_nan,axis=0)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=data,x="SS",y="TI",hue="PLAY_TYPE")
plt.title("Distribucin del tipo de play")
plt.show()
```
